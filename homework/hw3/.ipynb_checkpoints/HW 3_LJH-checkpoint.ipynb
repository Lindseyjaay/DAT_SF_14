#2 - Learn about label encoders at the following link and use what you learn to transform the features to numerical features.
http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

```python
label = preprocessing.LabelEncoder()
bank_data.job = label.fit_transform(bank_data.job)
bank_data.marital = label.fit_transform(bank_data.marital)
bank_data.education = label.fit_transform(bank_data.education)
bank_data.default = label.fit_transform(bank_data.default)
bank_data.housing = label.fit_transform(bank_data.housing)
bank_data.loan = label.fit_transform(bank_data.loan)
bank_data.contact = label.fit_transform(bank_data.contact)
bank_data.month = label.fit_transform(bank_data.month)
bank_data.poutcome = label.fit_transform(bank_data.poutcome)
bank_data.y = label.fit_transform(bank_data.y)
```

```python
bank_data
```

#3 - Build a decision tree model to predict whether a prospect will buy the product.

```python
cols = bank_data.columns[1:]

X = bank_data[cols]
y = bank_data.y

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print X_train
print y_train
print X_test
print y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
treeclass = DecisionTreeClassifier(random_state=1)
treeclass.fit(X_train, y_train)
```

```python
preds = treeclass.predict(X_test)


for i in preds:
    print i
```

```python
len(preds == y_test)/ len(y_test)
```

#4 - Evaluate the accuracy of your decision tree model using cross validation.

```python
from sklearn.cross_validation import cross_val_score

clf = DecisionTreeClassifier(random_state=0)
scores = cross_val_score(clf, X_test, y_test, cv=4)
print scores
```
